### Data Cleaning Section

```python
#Find Invalid Records

df.loc[(df['x']==0) | (df['y']==0) | (df['z']==0)]
```

```python
#Remove Invalid Records

drop_index = df.loc[
    (df['x']==0) |
    (df['y']==0) |
    (df['z']==0)
].index

df = df.drop(drop_index)
```

```python
df.describe()
```

### Category Analysis

```python
#Unique Values

print(df['cut'].unique())
print(df['color'].unique())
print(df['clarity'].unique())
```

### Encoding

```python
#Convert Categories to Numbers

df['color'] = df['color'].factorize()[0]
df['cut'] = df['cut'].factorize()[0]
df['clarity'] = df['clarity'].factorize()[0]
```

```python
df.head()
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df.head()
```

### Business Analysis

```python
#Price Per Carat

buy = df.copy()

buy['price_per_carat'] = buy['price'] / buy['carat']
```

```python
#Sort by Best Value

buy = buy.sort_values(
    by='price_per_carat',
    ascending=True
)
```

```python
#Best Value Diamond

print(f"Best value for money: {round(min(buy['price_per_carat']),2)}$")
print("Top 3 Diamonds")
buy.head(3)
```

### Additional Business Insights

#### Insight 1: Which Cut Type Has the Highest Average Price?

```python
df.groupby('cut')['price'].mean().sort_values(ascending=False)
```

###### Not all cuts are equally valuable. This shows which cut commands the highest average selling price.

#### Insight 2: Which Color Grade Has the Highest Average Price?

```python
df.groupby('color')['price'].mean().sort_values(ascending=False)
```

###### Find out whether premium color grades actually sell for higher prices.

#### Insight 3: Which Clarity Grade Has the Highest Average Price?

```python
df.groupby('clarity')['price'].mean().sort_values(ascending=False) 
```

##### Higher clarity generally means fewer imperfections and higher value.

#### Insight 4: Correlation Between Features and Price

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

#### Insight 5: Top 10 Most Expensive Diamonds

```python
df.nlargest(10, 'price')
```

##### Analyze characteristics of the most expensive diamonds.

#### Insight 6: Average Price by Carat Range

```python
df['carat_group'] = pd.cut(
    df['carat'],
    bins=[0,0.5,1,1.5,2,5],
    labels=['0-0.5','0.5-1','1-1.5','1.5-2','2+']
)

df.groupby('carat_group')['price'].mean()
```

##### How much does price increase as carat size increases?

#### Insight 7: Which Cut Is Most Popular?

```python
df['cut'].value_counts(normalize=True)*100
```

##### Shows customer preference distribution by cut quality.

#### Insight 8: Best Value-for-Money Category

```python
df['price_per_carat'] = df['price']/df['carat']

df.groupby('cut')['price_per_carat'].mean().sort_values()
```
